# Data 

In this part, I just try to explain the digit dataset ( sikitlearn default digit dataset) in many aspects like shape, targets, images and so on.

```python
#import essential modules
from sklearn.datasets import load_digits
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
digits = load_digits()
print(digits.keys())
```

```python
number_of_samples = digits.data.shape[0]
number_of_features = digits.data.shape[1]

print(f"number of samples : {number_of_samples}")
print(f"number of features : {number_of_features}")
```

```python
targets = digits.target_names
print(targets)
```


Example :

```python
print(f"sample[1] :\n {digits.data[0].reshape(8,8)} \n")
print(f"class : {digits.target[0]}")
plt.imshow(digits.data[0].reshape(8,8))
plt.show()
```

Looking from an overall perspective :

```python
fig = plt.figure(figsize=(7,7))

for i in range(49):
    ax = fig.add_subplot(7, 7, i + 1, xticks=[], yticks=[])
    ax.imshow(digits.images[i], cmap=plt.cm.binary, interpolation='nearest')
    ax.text(0, 7, str(digits.target[i]), bbox=dict(facecolor='blue', alpha=0.5))
```

```python
# according to 10th feature

for i in range(len(digits.target_names)):
    plt.hist(digits.data[digits.target==i,10])
    
plt.xlabel(digits.feature_names[10])
plt.title("distribution")
plt.show()
```

```python
# according to 5th feature

for i in range(len(digits.target_names)):
    plt.hist(digits.data[digits.target==i,5])
    
plt.xlabel(digits.feature_names[5])
plt.title("distribution")
plt.show()
```

```python
for i in range(len(digits.target_names)):
    plt.scatter(digits.data[digits.target==i,10],digits.data[digits.target==i,5])
    
plt.xlabel(digits.feature_names[10])
plt.ylabel(digits.feature_names[5])
plt.show()
```
